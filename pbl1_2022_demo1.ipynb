```python
import numpy as np
import cv2 as cv
from pbl1.imutils import imshow
  
```

```python
img = cv.imread('/content/pbl1/rainriver.bmp', cv.IMREAD_GRAYSCALE)
img = img.astype(np.float32)

[h, w] = img.shape
print("Width = ", w)
print("Height = ", h)

imshow(img, cmap="gray")
```

```python
img_res = np.zeros((h, w), dtype=np.float32)

alpha = 4
beta = -200

min_val = img.min()
max_val = img.max()
alpha = 1/(max_val-min_val)
beta = min_val/(max_val-min_val)

for y in range(0, h):
  for x in range(0, w):
    # img_res[y, x] = alpha*img[y, x] + beta
    img_res[y, x] = 255*(img[y, x] - min_val) / (max_val - min_val)

    # if (img_res[y, x] < 0):
    #   img_res[y, x] = 0
    # elif (img_res[y, x] > 255):
    #   img_res[y, x] = 255

img_res = img_res.clip(0, 255)
imshow(img_res, "gray")

print(img_res.min())
print(img_res.max())
```
